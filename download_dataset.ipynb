```python
from sklearn.datasets import fetch_lfw_people
import os, cv2

lfw = fetch_lfw_people(min_faces_per_person=20, resize=1.0, color=True)

print(f"✅ Loaded {lfw.images.shape[0]} images")
print(f"People found: {list(lfw.target_names)}")

os.makedirs("known_faces", exist_ok=True)

for idx, (image, label) in enumerate(zip(lfw.images, lfw.target)):
    name = lfw.target_names[label].replace(' ', '_')
    person_dir = f"known_faces/{name}"
    os.makedirs(person_dir, exist_ok=True)
    img_bgr = cv2.cvtColor(image.astype('uint8'), cv2.COLOR_RGB2BGR)
    cv2.imwrite(f"{person_dir}/{idx}.jpg", img_bgr)

print("✅ All images saved to known_faces/ folder!")
```
